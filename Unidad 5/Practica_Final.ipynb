```python
import matplotlib.pyplot as plt
from statistics import*
import pandas as pd
import math
from google.colab import drive
import seaborn as sns
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
```

```python
grupo1=pd.read_csv('grupo1.csv')
grupo2=pd.read_csv('grupo2.csv')
grupo3=pd.read_csv('grupo3.csv')
grupo1.describe()
grupo2.describe()
grupo3.describe()
```

```python
plt.hist(grupo1['Calificacion'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Calificaciones')
plt.ylabel('Frecuencia')
plt.title('Histograma de Calificaciones grupo1')
```

```python
plt.hist(grupo2['Calificacion'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Calificaciones')
plt.ylabel('Frecuencia')
plt.title('Histograma de Calificaciones grupo2')

```

```python
plt.hist(grupo3['Calificacion'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Calificaciones')
plt.ylabel('Frecuencia')
plt.title('Histograma de Calificaciones grupo3')
```

```python
x = grupo1[['HorasEstudio']]
y = grupo1['Calificacion']
model = LinearRegression()
model.fit(x, y)
pendiente = model.coef_[0]
intercepto = model.intercept_
print("Ecuación de la línea de regresión:")
print(f"Calificaciones = {pendiente:.2f} * Horas de Estudio + {intercepto:.2f}")
plt.scatter(grupo1['HorasEstudio'], grupo1['Calificacion'], color='blue', label='Datos')
plt.plot(grupo1[['HorasEstudio']], model.predict(x), color='red', label='Línea de regresión')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificaciones')
plt.title('Regresión Lineal: Horas de Estudio vs Calificaciones Grupo1')
plt.legend()
plt.show()

coef_corr, _ = pearsonr(grupo1['HorasEstudio'], grupo1['Calificacion'])
print("Coeficiente de correlación de Pearson:",coef_corr)
```

```python
plt.figure(figsize=(6,4))
sns.set(style='whitegrid')
sns.regplot(x=grupo1[['HorasEstudio']], y=grupo1['Calificacion'], scatter=True,color='blue')
plt.title('Diagramna de dispersión con linea de regresion del grupo1')
plt.xlabel('Horas de estudio')
plt.ylabel('calificaciones')
plt.grid(True)
plt.show()
```
