```python
#Libreria de python para generar numeros reales aleatrorios
import random
#Libreria de python para graficar
import matplotlib.pyplot as plt
from statistics import*
```

```python
fabrica1=[]
for i in range(100):
  numero_aleatorio = round (random.uniform(50,100), 2)
  fabrica1.append(numero_aleatorio)
print (fabrica1)
```

```python
fabrica2=[]
for i in range(100):
  numero_aleatorio2 = round (random.uniform(50,100), 2)
  fabrica2.append(numero_aleatorio2)
print (fabrica2)
```

```python
plt.hist(fabrica1)
plt.title("Producción diaria de la fabrica 1")
```

```python
plt.hist(fabrica2)
plt.title("Producción diaria de la fabrica 2")
```

```python
media1 = mean(fabrica1)
mediana1=median(fabrica1)
moda1=mode(fabrica1)
print("valores de la fabrica 1")
print("La media es: ",media1)
print("la moda es: ",moda1)
print("La mediana es: ", mediana1)
```

```python
media = mean(fabrica2)
mediana=median(fabrica2)
moda=mode(fabrica2)
print("valores de la fabrica 2")
print("La media es: ",media)
print("la moda es: ",moda)
print("La mediana es: ", mediana)
```

```python
#generar sublot de un renglon por dos columnas
fig,(ax1, ax2)= plt.subplots(nrows=1,ncols=2, figsize=(10,5))
ax1.hist(fabrica1)
ax2.hist(fabrica2)
plt.show()
```

### Maximo, minimo y rango

```python
maximo = max (fabrica1)
minimo = min(fabrica1)
rango = maximo - minimo
media1 = mean (fabrica1)
mediana1= median(fabrica1)
moda1= mode(fabrica1)
print("estadisticas de la fabrica 1: maximo", maximo," minimo", minimo," rango", rango," media", media1, " mediana", mediana1," moda", moda1)
```

```python
maximo = max (fabrica2)
minimo = min(fabrica2)
rango = maximo - minimo
media = mean (fabrica2)
mediana= median(fabrica2)
moda= mode(fabrica2)
print("estadisticas de la fabrica 1: maximo", maximo,"\n minimo", minimo," \nrango", rango," \nmedia", media, " \n mediana", mediana," \n moda", moda)
```

###Dispersion de datos

```python
Desviacionestandar_fabrica1=stdev(fabrica1)
print("La desviacion estandar de la fabrica 1 es: ",Desviacionestandar_fabrica1)
```

```python
Desviacionestandar_fabrica2=stdev(fabrica2)
print("La desviacion estandar de la fabrica 2 es: ",Desviacionestandar_fabrica2)
```

```python
plt.figure(figsize=(11,8))
plt.hist(fabrica1, bins=20, alpha = 0.2, color='blue', edgecolor='black')
plt.axvline(media1, color='r', linestyle='dashed', linewidth=2,label='Media')
plt.axvline(mediana1, color='purple', linestyle='dashed', linewidth=2,label='Mediana')
plt.axvline(moda1, color='g', linestyle='dashed', linewidth=2,label='Moda')
plt.axvline(media1 - Desviacionestandar_fabrica1, color='y', linestyle='dashed', linewidth=2,label=f'Desviacion estandar={Desviacionestandar_fabrica1:.2f}')
plt.axvline(media1 + Desviacionestandar_fabrica1, color='y', linestyle='dashed', linewidth=2)
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title("Producción diaria de la fabrica 1")
plt.legend()
plt.show()
```
