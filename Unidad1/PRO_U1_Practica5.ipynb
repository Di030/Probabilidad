#Tablas de frecuencias

```python
edades=[]
for i in range (100):
  num_aleatorio=random.randint(18,70)
  edades.append(num_aleatorio)
print(edades)
```

```python
df_edades= pd.DataFrame(edades, columns=["edades"])
df_edades
```

```python
tabla_frecuencias=df_edades['edades'].value_counts().reset_index()
tabla_frecuencias.columns = ['Edad','Frecuencia']
tabla_frecuencias = tabla_frecuencias.sort_values(by = "Edad")
tabla_frecuencias
```

```python
tabla_frecuencias["Frec.Relativa"]=tabla_frecuencias['Frecuencia']/100
tabla_frecuencias
```

```python
tabla_frecuencias["frec.porcentual"]=tabla_frecuencias["Frecuencia"]/100*100
tabla_frecuencias
```

```python
tabla_frecuencias["Frecuencia acumulada"]=tabla_frecuencias["Frecuencia"].cumsum()
tabla_frecuencias
```

#Regla de sturges
1.- Es una regla que sirve para calucular el numero de clases o intervalos idóneo, en los que se debe de dividir un conjunto de datos

$
c=1+ log_2(N)
$

```python
print(edades)
```

```python
plt.hist(edades, alpha=0.5, rwidth=0.95)
plt.title("Histograma de edades basado en valores estandar de matplotlib")
plt.xlabel("edades")
plt.ylabel("Frecuencias")
```

```python
df_edades_sturges = pd.DataFrame({"Edad": edades})
df_edades_sturges
```

```python
n = len(edades)
k = int(1+math.log2(n))
print("el valor de n= ",n," valor de num de intervalos= ", k)
```

```python
tabla_frec_sturges = pd.cut(df_edades_sturges['Edad'], bins=k).value_counts().reset_index()
tabla_frec_sturges
```

```python
tabla_frec_sturges.columns=['Clase','Frecuencia']
tabla_frec_sturges
```

```python
tabla_frec_sturges=tabla_frec_sturges.sort_values(by ='Clase')
tabla_frec_sturges
```

```python
tabla_frec_sturges['Frec.relativa']=tabla_frec_sturges['Frecuencia']/n
tabla_frec_sturges['Frec.porcentual']=tabla_frec_sturges['Frecuencia']/n*100
tabla_frec_sturges['Frec.acumulada']=tabla_frec_sturges['Frecuencia'].cumsum()
tabla_frec_sturges
```

```python
plt.hist(df_edades_sturges['Edad'], bins=k, alpha=0.5,rwidth=0.95)
plt.title("Histograma basado en la regla de Sturges")
plt.xlabel("edades")
plt.ylabel("Frecuencias")
```
